```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np


file = 'cat.jpeg'
# Baca gambar
gambar1 = mpimg.imread(file)

# Menampilkan informasi gambar
info = mpimg.imread(file)
print(info)

# Menampilkan gambar dengan imtool
plt.figure()
plt.imshow(gambar1)
plt.title('imtool')
plt.show()

# Menampilkan gambar dengan subplot imagesc dan imshow
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(gambar1)
plt.title('imagesc')
plt.subplot(1, 2, 2)
plt.imshow(gambar1)
plt.title('imshow')

# Memisahkan saluran warna (R, G, B)
R = gambar1[:, :, 0]
G = gambar1[:, :, 1]
B = gambar1[:, :, 2]

# Membuat saluran warna terpisah (redOnly, greenOnly, blueOnly)
redOnly = np.zeros_like(gambar1)
redOnly[:, :, 0] = R

greenOnly = np.zeros_like(gambar1)
greenOnly[:, :, 1] = G

blueOnly = np.zeros_like(gambar1)
blueOnly[:, :, 2] = B

# Menampilkan saluran warna secara terpisah
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.imshow(gambar1)
plt.title('Original')
plt.subplot(2, 2, 2)
plt.imshow(redOnly)
plt.title('Red Only')
plt.subplot(2, 2, 3)
plt.imshow(greenOnly)
plt.title('Green Only')
plt.subplot(2, 2, 4)
plt.imshow(blueOnly)
plt.title('Blue Only')

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np


file = 'cat.jpeg'
# Baca gambar
gambar1 = mpimg.imread(file)

# Menampilkan informasi gambar
info = mpimg.imread(file)
print(info)

# Menampilkan gambar dengan imtool (OpenCV)
plt.imshow(gambar1)
plt.title('imtool')
plt.show()

# Menampilkan gambar dengan subplot imagesc dan imshow
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(gambar1)
plt.title('imagesc')
plt.subplot(1, 2, 2)
plt.imshow(gambar1)
plt.title('imshow')

# Konversi NumPy array menjadi objek gambar PIL
gambar1_pil = Image.fromarray(gambar1)

# Memisahkan saluran warna (R, G, B)
R, G, B = gambar1_pil.split()


# Membuat saluran warna terpisah (redOnly, greenOnly, blueOnly)
redOnly = Image.merge('RGB', (R, R, R))
greenOnly = Image.merge('RGB', (G, G, G))
blueOnly = Image.merge('RGB', (B, B, B))

# Menampilkan saluran warna secara terpisah
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.imshow(gambar1)
plt.title('Original')
plt.subplot(2, 2, 2)
plt.imshow(redOnly)
plt.title('Red Only')
plt.subplot(2, 2, 3)
plt.imshow(greenOnly)
plt.title('Green Only')
plt.subplot(2, 2, 4)
plt.imshow(blueOnly)
plt.title('Blue Only')

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np


file = 'cat.jpeg'
# Baca gambar
gambar1 = mpimg.imread(file)

# Menampilkan informasi gambar
info = mpimg.imread(file)
print(info)

# Menampilkan gambar dengan Matplotlib
plt.imshow(gambar1)
plt.title('imtool')
plt.show()

# Menampilkan gambar dengan subplot imagesc dan imshow
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(gambar1)
plt.title('imagesc')
plt.subplot(1, 2, 2)
plt.imshow(gambar1)
plt.title('imshow')

# Konversi NumPy array menjadi objek gambar PIL
gambar1_pil = Image.fromarray(gambar1)

# Memisahkan saluran warna (R, G, B)
R, G, B = gambar1_pil.split()

# Membuat saluran warna terpisah (redOnly, greenOnly, blueOnly)
redOnly = Image.merge('RGB', (R, Image.new('L', R.size, 0), Image.new('L', R.size, 0)))
greenOnly = Image.merge('RGB', (Image.new('L', G.size, 0), G, Image.new('L', G.size, 0)))
blueOnly = Image.merge('RGB', (Image.new('L', B.size, 0), Image.new('L', B.size, 0), B))

# Menampilkan saluran warna secara terpisah
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.imshow(gambar1)
plt.title('Original')
plt.subplot(2, 2, 2)
plt.imshow(redOnly)
plt.title('Red Only')
plt.subplot(2, 2, 3)
plt.imshow(greenOnly)
plt.title('Green Only')
plt.subplot(2, 2, 4)
plt.imshow(blueOnly)
plt.title('Blue Only')

plt.show()
```
